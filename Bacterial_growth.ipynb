```python
import numpy as np
import matplotlib.pyplot as plt

# Time points
time = np.arange(0, 10, 1)  # 10 time points

# Bacterial growth data (sample)
bacterial_growth = np.array([1, 2, 4, 8, 16, 32, 64, 128, 256, 512])

# Plot bacterial growth over time
plt.plot(time, bacterial_growth, marker='o', linestyle='-')

plt.xlabel('Time (hours)')
plt.ylabel('Bacterial Count')
plt.title('Bacterial Growth Over Time')
plt.grid(True)
plt.show()
```
