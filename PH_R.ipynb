```python
import numpy as np
import matplotlib.pyplot as plt

# Titration data (example)
volume = np.linspace(0, 50, 100)  # Volume of titrant (in mL)
pH = -np.log10(volume/100)        # Example pH values, assuming a strong acid and strong base titration

# Plot titration curve
plt.plot(volume, pH, color='green')

plt.xlabel('Volume of Titrant (mL)')
plt.ylabel('pH')
plt.title('Titration Curve')
plt.grid(True)
plt.show()
```
